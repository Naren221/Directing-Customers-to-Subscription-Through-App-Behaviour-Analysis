We will check for the distribution of dayofweek, hour, age, numscreens, minigame, used_premimum_feature, enrolled.
Of these the numeric features are hour ,age and numscreens.
And the categoric variables are dayofweek, minigame, used_premium feature and enrolled.

#### Univariate Analysis

```python
Numeric_vars = ['hour','age','numscreens']
Categoric_vars = ['dayofweek', 'minigame', 'used_premium_feature', 'enrolled']
```

```python
title_font = {'family':'arial', 'color':'darkred', 'weight':'bold', 'size':15}
axis_font = {'family':'arial', 'color':'darkblue','weight':'bold', 'size':15}
```

```python
numeric_axis_name = ['Hour of the Day', 'Age', 'Number Of Screens']
```

```python
Categoric_axis_name = ['Day of Week', 'MiniGame', 'used Premium Feature', 'Enrolled']
```

```python
for i,z in list(zip(Numeric_vars, numeric_axis_name)):
    plt.figure(figsize = (8,6), dpi = 80)
    sns.distplot(df[i], hist_kws = dict(linewidth =1, edgecolor = 'k'), bins = 20)
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict = axis_font)
    plt.ylabel('Density', fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```

```python
for i , z in list(zip(Categoric_vars, Categoric_axis_name)):
    fig,ax = plt.subplots(figsize = (8,6))
    
    observation_vlaues = list(df[i].value_counts().index)
    total_observation_values = list(df[i].value_counts())
    
    ax.pie(total_observation_values, labels = observation_vlaues , autopct = '%1.1f%%', startangle = 110, labeldistance= 1.1)
    ax.axis('equal')
    
    plt.title((i +"(" + z +")"),fontdict= title_font)
    plt.legend()
    plt.show()
    
    
```
